```python
import pandas as pd
```

```python
df1=pd.read_csv('sdoh_data.csv')
df2=pd.read_csv('SVI_USCounty_2020.csv')

# Drop columns with 'Unnamed' in their names
df2 = df2.loc[:, ~df2.columns.str.contains('^Unnamed')]
#print(df2.head())
#print(df1.head())

merged_df = pd.merge(df1, df2, how='left', on='FIPS')
merged_df = merged_df.drop(columns=['ST_ABBR'])
merged_df
```

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

# Geospatial data for US counties
county_shapes = gpd.read_file("https://raw.githubusercontent.com/plotly/datasets/master/geojson-counties-fips.json")

# Ensure FIPS codes are strings with leading zeros
merged_df['FIPS'] = merged_df['FIPS'].astype(int).astype(str).str.zfill(5)
# county_shapes['id'] = county_shapes['id'].str.zfill(5)
county_shapes['FIPS']=county_shapes['STATE']  + county_shapes['COUNTY']
# print(county_shapes['geometry'])

# Load state-level geospatial data
state_shapes = gpd.read_file("ne_110m_admin_1_states_provinces.shp")
state_shapes = state_shapes[state_shapes['iso_a2'] == "US"]  # Filter for US only
#print(state_shapes)

# Merge your dataset with geospatial data
merged = county_shapes.merge(merged_df, left_on="id", right_on="FIPS", how="left")
#print(merged)

# Add a column for state FIPS (first two digits of county FIPS)
merged["state_fips"] = merged["id"].str[:2]

# Plotting the map 
fig, ax = plt.subplots(1, 1, figsize=(20, 12))
merged.boundary.plot(ax=ax, linewidth=0.5, color="gray") #County Boundaries

merged.plot(column='Value', cmap='OrRd', linewidth=0.5, ax=ax, edgecolor='1.0', legend=True)

# Overlay state boundaries for clarity
state_shapes = merged.dissolve(by="state_fips")  # Group counties into states
state_shapes.boundary.plot(ax=ax, linewidth=1, color="black") #State Boundaries

# Add titles and labels
ax.set_title("County-Level SVI", fontdict={'fontsize': '20', 'fontweight' : '3'})
ax.axis("off")

# Focus the map on the continental US for better clarity
ax.set_xlim([-130, -65])
ax.set_ylim([23, 50])

plt.show()
```

```python
# Group by 'state' and calculate the mean SVI score for each state
state_avg_svi = merged_df.groupby('State')['Value'].mean().reset_index()

# Rename columns for clarity
state_avg_svi.columns = ['State', 'Avg_SVI']

# Display the result
print(state_avg_svi)
```

```python
# List of theme columns to include in the calculation
theme_columns = ['FIPS', 'SPL_THEME1', 'SPL_THEME2', 'SPL_THEME3', 'SPL_THEME4']

# Group by 'state' and calculate the mean for each theme column
state_avg_themes = merged_df.groupby('State')[theme_columns].mean().reset_index()

# Display the result
state_avg_themes
```

```python
# Merge on the 'state' column
merged_df2 = pd.merge(state_avg_svi, state_avg_themes, on='State')

# Display the merged DataFrame
merged_df2
```

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
data = {
    "State": ["AK", "AL", "AR", "AZ", "CA", "CO", "CT", "DC", "DE", "FL"],
    "FIPS": [2140.266667, 1067.000000, 5075.000000, 4013.866667, 6058.000000, 8062.234375, 9008.000000, 11001.000000, 10004.000000, 12067.910448],
    "SPL_THEME1": [2.558487, 3.130284, 2.933519, 3.330680, 2.902131, 2.383169, 1.992200, 2.356100, 2.194750, 3.154379],
}
df = pd.DataFrame(data)

# Load county-level geospatial data
county_shapes = gpd.read_file("https://raw.githubusercontent.com/plotly/datasets/master/geojson-counties-fips.json")

# Ensure FIPS codes are strings with leading zeros
df['FIPS'] = df['FIPS'].astype(int).astype(str).str.zfill(5)
county_shapes['id'] = county_shapes['id'].str.zfill(5)

# Merge your dataset with geospatial data
merged = county_shapes.merge(df, left_on="id", right_on="FIPS", how="left")

# Add a column for state FIPS (first two digits of county FIPS)
merged["state_fips"] = merged["id"].str[:2]

# Plotting the map
fig, ax = plt.subplots(1, 1, figsize=(20, 12))

# Plot counties, coloring them based on their state FIPS code
merged.plot(column='state_fips', cmap='tab20', linewidth=0.1, ax=ax, edgecolor='white', legend=False)

# Overlay state boundaries for clarity
state_shapes = merged.dissolve(by="state_fips")  # Group counties into states
state_shapes.boundary.plot(ax=ax, linewidth=1, color="black")

# Add titles and labels
ax.set_title("Counties Grouped by State", fontdict={'fontsize': '20', 'fontweight': '3'})
ax.axis("off")

plt.show()
```
